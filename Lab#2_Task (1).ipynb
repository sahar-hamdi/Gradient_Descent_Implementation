```python
x = np.arange(200 , )
delta = np.random.uniform(-50 , 30 , size = (200,))
y = 0.4 * x + 3 + delta
```

**Split Data into test and train**

```python
x_train , x_test , y_train , y_test = train_test_split(x , y , test_size = 0.20 , random_state = 0)
```

**Mean Squared Error implementation**

**Gradient Descent Implementation**

```python
#intialize w , b , num_iterations and learning_rate
w = np.random.randn()
b = np.random.randn()
alpha = 0.0001
num_iterations = 1000
```

```python
weights = []
lose = []
bais = []
m = int(0.8 * len(x))
dw = 0 
db = 0
for i in range (1000):
    y_pred = w * x_train + b 
    dw = (1/m) * np.sum((y_pred - y_train) * x_train)
    db = (1/m) * np.sum(y_pred - y_train)
    #Update values of w and b.  
    w = w - alpha * dw
    b = b - alpha * db
    weights.append(w)
    bais.append(b)
    if i % 100 == 0:
        loss =  np.mean((y_pred - y_train)**2)
        print(f"Iteration {i}: Loss = {loss}")
```

```python
print("Weights = " , weights)
print("Biases = " , bais)
```

**Calculate Accuracy**

```python
y_prediction = w * x_test + b
accuracy = np.mean((y_prediction - y_test)**2)
print("Accuracy = ", accuracy)
```
