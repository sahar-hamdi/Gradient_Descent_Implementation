```python
def MSE(x, y, w, b):
   
    m = x.shape[0] 
    cost = 0
    
    for i in range(m):
        loss = w * x[i] + b
        cost = cost + (loss - y[i])**2
    total_cost = 1 / (2 * m) * cost

    return total_cost
```

```python
def gred(x, y, w, b): 

    # Number of training examples
    m = x.shape[0]    
    dw = 0
    db = 0
    
    for i in range(m):  
        cost = w * x[i] + b 
        dw_i = (cost - y[i]) * x[i] 
        db_i = cost - y[i] 
        db += db_i
        dw += dw_i 
    dw /= m 
    db /= m 
        
    return dw, db
```

```python
def gradient_descent(x_train , y_train , w , b , alpha , num_iterations , MSE , gred):
    loss = []
    w_b_values = []
    for i in range (num_iterations):
        dw , db = gred(x , y , w , b)
        #caculate Gradient Descent and update values of w and b. 
        w = w - alpha * dw
        b = b - alpha * db
        if i % 1000 == 0 :
            loss.append(MSE(x , y , w , b))
            w_b_values.append([w , b])
    return w , b , loss , w_b_values
```
