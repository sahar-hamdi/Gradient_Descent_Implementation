```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn as skl
from sklearn.model_selection import train_test_split
```

```python
x = np.arange(200)
delta = np.random.uniform(-50 , 30 , size = (200,))
y = 0.4 * x + 3 + delta
```

```python
plt.scatter(x , y)
plt.show()
```

**Split Data into test and train**

```python
x_train , x_test , y_train , y_test = train_test_split(x , y , test_size = 0.20 , random_state = 0)
```

**Mean Squared Error implementation**

```python
def MSE(y , y_pred):
    return np.square(np.subtract(y_pred , y)).mean()
```

**Gradient Descent Implementation**

```python
#intialize w , b , num_iterations and learning_rate
w = 0 
b = 0 
alpha = 000.1
num_iterations = 1000
```

```python
def gradient_descent(x_train , y_train , w , b , alpha , num_iterations = 1000):
    weights = []
    lose = []
    bais = []
    for i in range (1000):
        y_pred = w * x + b 
        loss_func = MSE(y , y_pred)
        lose.append(loss_func)
    #caculate Gradient Descent and update values of w and b. 
    w = w - alpha * (np.mean(y_pred - y) * x)
    b = b - alpha * (np.mean(y_pred - y))
    weights.append(w)
    bais.append(b)
    return weights , lose , bais
```

```python
lose , weights , bais = gradient_descent(x_train , y_train , w , b , alpha , num_iterations)
```

```python
print("Weight = " , weights[-1]) 
print("bais = " , bais[-1]) 

#Prediction
y_prediction = weights[-1] * x_test + bais[-1]
```

**Calculate Accuracy**

```python
accuracy = MSE(y_test , y_prediction)
print("Accuracy = ", accuracy)
```
